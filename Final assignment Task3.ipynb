##### 6. Peak Hours of Sales

```python
#Etract the hours from Time

print(final_df.columns)

final_df["date"]= pd.to_datetime(final_df["date"].astype(str)+ ' ' + final_df["time"].astype(str), errors="coerce")
```

```python
print(final_df.columns.tolist())
```

```python
date_col= [col for col in final_df.columns if col.lower() == "date"][0]
time_col= [col for col in final_df.columns if col.lower() == "time"][0]
```

```python
final_df["datetime"]= pd.to_datetime(final_df[date_col].astype(str)+ " " + final_df[time_col].astype(str),errors="coerce")
                                     
```

```python
final_df["hour"]= final_df["datetime"].dt.hour
```

```python
print(final_df[["datetime","hour"]].head())
```

```python
# Group by hour
orders_by_hour= final_df.groupby("hour")["order_id"].nunique()
```

```python
orders_by_hour
```

```python
#Let sort to get the peak hours

peak_hours= orders_by_hour.sort_values(ascending=False)
```

```python
print("Peak Hours of Sales: \n", peak_hours.head(10))
```

```python
orders_by_hour.plot(kind="bar", figsize=(10,5), color= "skyblue")
plt.title("Orders by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Orders")

plt.show()
```

##### 7. Find the Total sales per weekdays

###### The below is the breakdown of sales per each days

```python
final_df["weekday"]= final_df["datetime"].dt.day_name()
```

```python
sales_per_weekday =(final_df["quantity"]*final_df["price"]).groupby(final_df["weekday"]).sum()
```

```python
print(sales_per_weekday)  #to indicate each sale perday of the week
```

###### To Calculate the Best day of the week

```python
best_day = sales_per_weekday.idxmax()
best_sales = sales_per_weekday.max()
print("Best Sales Day: ", best_day, "With Sales Of", best_sales)
```

```python
best_day = sales_per_weekday.idxmax()
print("Best Day: ", best_day, "With Sales On", best_day)
```

```python
sales_per_weekday= sales_per_weekday.reindex(["Monday","Tuesday","Wednesday","Thursday","Saturday","Sunday"])
sales_per_weekday.plot(kind="bar", figsize=(8,5), color= "Orange")
plt.title("Total Sales by Weekda")
plt.xlabel("Day of the week")
plt.ylabel("Total Sales")

plt.show()
```

###### 8. Top 5 Bestselling Pizzas

```python
#Let sum up all quantity sold for each pizza type

top5_pizzas = final_df.groupby("pizza_type_id")["quantity"].sum().sort_values(ascending= False).head(5)
print("Top 5 Bestselling Pizzas: \n", top5_pizzas)
```

```python

top5_pizzas_name = final_df.groupby("name")["quantity"].sum().sort_values(ascending= False).head(5)
print("Top 5 Bestselling Pizzas(With Names): \n", top5_pizzas_name)
```

```python

top5_pizzas_name.plot(kind="bar", figsize=(8,5), color= "Green")
plt.title("Top 5 Bestselling Pizzas")
plt.xlabel("Pizza Name")
plt.ylabel("Quantity Sold")

plt.show()
```

##### 9. Sales By Months

```python
#Let extract the Month

final_df["Month"] = final_df["datetime"].dt.month_name()
```

```python
# Let Calculate sales per Months

sales_by_month = (final_df["quantity"]*final_df["price"]).groupby(final_df["Month"]).sum()
print(sales_by_month)
```

#Let identify the strongest month

best_month= sales_by_month.idxmax()
best_sales= sales_by_month.max()
print("Best Month: ", best_month, "With Sales of",best_sales)

```python
month_order = ["January","February","March","April","May","June","July","August","September","October","November","December"]
sales_by_month= sales_by_month.reindex(month_order)
sales_by_month.plot(kind="line", marker= "o",figsize=(10,5), color= "Purple")
plt.title("Total Sales by Month")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.show()
```

###### 10. Underperming Pizzas on the Menu
